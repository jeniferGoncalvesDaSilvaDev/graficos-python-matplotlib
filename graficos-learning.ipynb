```python
#graficos de quantidade de marmitas, vendidas diariamente, em 5 dias, dentro de uma semana, de 3 empresas diferentes
#Empresa: Beta Log
from matplotlib import pyplot as plt
numero_marmitas_beta_log=[12,15,7,10,10]
dias_beta_log=[1,2,3,4,5]
#tamanho do grafico
figura=plt.figure(figsize=(30,4))
#titulos na figura, que vai ser o titulo dos graficos
figura.suptitle('Análise de gráficos de vendas de marmitas')
#para plotar o primeiro gráfico
#subplot, pega o numero de linhas ou colunas, posiçao e eixo na figura que os graficos, vao ser plotados
figura.add_subplot(131)
plt.xlabel('Dias')
plt.ylabel('Número de marmitas vendidas para Beta Log')
plt.plot(dias_beta_log, numero_marmitas_beta_log,label='Número de marmitas vendidas')

plt.legend()
plt.title('Gráfico de marmitas vendidas da empresa Beta Log')
#Empresa Tag Log
numero_marmitas_tag_log=[20,12,15,8,17]
dias_tag_log=[1,2,3,4,5]
#nos gráficos posteriores, tendo ja definido as dimensoes da figura, agora é so subplotar!
figura.add_subplot(133)
plt.xlabel('Dias')
plt.ylabel('Número de marmitas para Tag Log')
plt.plot(dias_tag_log,numero_marmitas_tag_log, label='Número de mamritas vendidas')
plt.legend()
plt.title('Gráfico de marmitas vendidas da empresa Tag Log')
#Empresa: Pescados
marmitas_pescados=[10,9,14,5,15]
dias_pescados=[1,2,3,4,5]
figura.add_subplot(132)
plt.xlabel('Dias')
plt.ylabel('Número de marmitas vendidas para Pescados')
plt.plot(dias_pescados,marmitas_pescados,label="Número de marmitas vendidas")
plt.legend()
plt.title("Gráfico de marmitas vendidas da empresa Pescados")
#salvando os graficos
plt.savefig('grafico-vendas-marmitas.png')
plt.show(figura)
```
